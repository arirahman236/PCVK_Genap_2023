```python
img = cv.imread('/content/drive/MyDrive/gambar/soal3.jpg')
img = cv.resize(img, (0,0), fx=0.10, fy=0.10)
plt.imshow(img)
```

```python
img = cv.imread('/content/drive/MyDrive/gambar/soal3.jpg')
tinggi = img.shape[0]
lebar = img.shape[1]
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

#Membuat garis horizontal di tengah image
for y in range (lebar):
  img[int((tinggi)/2),y] = [255, 255, 255]
final_frame = cv.hconcat((img))
for x in range (tinggi):
  img[x, int(lebar/2)] = [255, 255, 255]
final_frame = cv.hconcat((img))
cv2_imshow(final_frame)
```

```python
img = cv.flip(cv.cvtColor(img,cv.COLOR_BGR2RGB),0)

fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111)
ax.imshow(img)
```
